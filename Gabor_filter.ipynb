```python
import numpy as np
import cv2
import pandas as pd
from matplotlib import pyplot as plt
#img = cv2.imread('BSE_Image.jpg')
img = cv2.imread('synthetic.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
 
plt.imshow(img, cmap='gray')
```

```python
ksize=5
sigma=3
theta=1*np.pi/4
lamda=1*np.pi/4
gamma=0.5
phi=0
fig = plt.figure(figsize=(20, 20))   
kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, 0, ktype=cv2.CV_32F)
kernel_resized=cv2.resize(kernel,(400,400))
fig.add_subplot(2, 2, 1,xlabel='kernel')
plt.imshow(kernel_resized, cmap='gray')
 

#Now filter the image and add values to a new column 
fimg = cv2.filter2D(img, cv2.CV_8UC3, kernel)
fig.add_subplot(2, 2, 2,xlabel='filtered_image')
plt.imshow(fimg, cmap='gray')
    
```
